```python
#Created the data
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
pn=["Charlie Liu","Vladimir Kryazh","Sergei Mochtchenkov","Sergei Mochtchenkov","Merisa Lee","Ramu","Seetha","Varun","Sandeep","Himaja"]
data={"names":pn,
     "Years of experience":np.random.randint(0,15,size=10),
     "Got promotion or not":["Y","N","N","Y","N","Y","N",np.nan,np.nan,"N"],
    "remarks":np.random.choice(["Y","N"],size=10),
     "Salary":np.random.randint(0,1000000,size=10)}
df=pd.DataFrame(data)
print(df)
```

```python
#Checking for nan values
df.isnull().sum()
```

```python
# fill nan values with Y
df["Got promotion or not"].fillna("Y",inplace=True)
df
```

```python
#drop remark column
df.drop("remarks",axis=1,inplace=True)
```

```python
df
```

```python
#incremented the salary
df["increment"]=df["Salary"]+5000
        
```

```python
#Multiplying the salary
df["increment"]=df["Salary"]*5
```

```python
df
```

```python
#substracting  the salary
df["increment"]=df["Salary"]-5000
```

```python
df
```

```python
#Divided  the salary
df["increment"]=df["Salary"]//50
```

```python
plt.bar(list(df["names"]),list(df["increment"]))
plt.xlabel("names")
plt.ylabel("increment")
plt.show()
```
